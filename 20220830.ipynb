### 計算BMI並判斷體位

BMI的算法：
$ bmi = \frac{體重(公斤)}{身高(公尺)^2} $

$$  bmi = \frac{體重(公斤)}{身高(公尺)^2} $$

|bmi|說明|
|:--|:--|
|bmi < 18|體重過輕|
|18 < bmi < 24|健康體位|
|bmi > 24|體重過重|

````python
w = 80
h = 1.83
````

````python
bmi = w / h**2
````

````python
bmi
````

````python
if bmi < 18:
  print("體重過輕")
elif 18 < bmi < 24:
  print("健康體位")
else:
  print("體重過重")
````

## 例外處理

```python
try:
  程式區塊
except Exception as e:
  例外處理程式區塊
```

````python
def add(a, b):
  print(f"{a} + {b} = {a+b}")

def sub(a, b):
  print(f"{a} - {b} = {a-b}")

def mul(a, b):
  print(f"{a} x {b} = {a*b}")

def div(a, b):
  print(f"{a} / {b} = {a/b}")
````

````python
fns = [add, div, sub, mul]
````

````python
a = 1
b = 0

for fn in fns:
  fn(a, b)
````

````python
a = 1
b = 0

for fn in fns:
  try:
    fn(a, b)
  except Exception as e:
    print(e.args)
````

### 拆分訓練集跟測試集



````python
from sklearn.model_selection import train_test_split
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
````

### 使用 scikit-learn 做機器學習

````python
# 1. 載入模型
# from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
````

````python
# 2. 建立模型
# clf = SVC()
clf = RandomForestClassifier()
````

````python
# 3. 訓練模型
clf.fit(X_train, y_train)
````

````python
# 計算準確度
clf.score(X_test, y_test)
````

````python
# 4. 使用模型
clf.predict(X_test)
````
